```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, r2_score
```

```python
df = pd.read_csv("student_scores.csv")
```

```python
x = df[['Hours']]
y = df['Scores']
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

```python
# Prediction
prediction = lr.predict(X_test.iloc[2].values.reshape(1, 1))
```

```python
# Plotting the best fit line
plt.scatter(df['Hours'], df['Scores'])
plt.plot(X_test, lr.predict(X_test), color='red')
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.show()
```

```python
Y_pred = lr.predict(X_test)
r2 = r2_score(Y_test, Y_pred)
print("R2 Score of linear regression:", r2)
```

```python
# Logistic Regression
X = df[['Hours']]  # Features
df['Pass'] = (df['Scores'] > 50).astype(int)
y = df['Pass']# Target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
classifier = LogisticRegression(multi_class='ovr')  # Explicitly set the multi_class parameter
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
